```python
#importando a biblioteca pandas

import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Cusro_Python_Pandas_Digital_Innovation_master/datasets/Gapminder.csv", error_bad_lines=False, sep=";")
```

```python
#Visualizando as 5 primeiras linhas com o metodo head

df.head()
```

```python
#Abre-se chave {} para criar um dicionário

df = df.rename(columns={"country":"Pais", "continent": "Continente", "year":"Ano", "lifeExp":"Expec_vida", "pop":"Populacao", "gdpPercap":"PIB"})
```

```python
#Retornar apenas os nomes das colunas
#Abre-se colchetes para visualizar uma lista ou criar uma lista

df.columns
```

```python
#Como saber o tipo de dado em cada coluna
#object = string

df.dtypes
```

```python
#Visualizar os ultimos dados do arquivo - tail. Se não especificar, sempre ira retornar as ultimas 5 linhas

df.tail(3)
```

```python
#Calculos estatisticos automaticos que o Pandas faz com o metodo "describe"
#mean = média; std = desvio padrao; min = valor minimo; 25%, 50%(mediana) e 75% = quartis; max = valor maximo

df.describe()
```

```python
#Como retornar valores unicos dos dados de uma coluna
# nan quer dizer que há valores nulos em algumas linhas

df["Continente"].unique()
```

```python
#Como retornar valores apenas de um continente especifico
#Criar uma variavel com o nome do continente
#loc = localizacao
#sinal de igual em Python é para atribuir e o sinal com dois iguais "==" significa o sinal de igualdade

Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```python
#MOstra que dentro da variavel Oceania só tem o continente Oceania

Oceania["Continente"].unique()
```

```python
from pandas.core import groupby
#Para agrupar dados usa-se o "groupby"
#Nesse caso estamos pegando o grupo "Continente" e depois pegando só o grupo "Pais" e fazer uma contagem dos paises por continente
# nunique conta todos os paises ou o valor especificado

df.groupby("Continente")["Pais"].nunique()
```

```python
#Para cada ano, qual a expectativa de vida média em cada país nos dados?

df.groupby("Ano")["Expec_vida"].mean()
```

```python
#Calculando a media do PIB total

df["PIB"].mean()
```

```python
#Somando o PIB total

df["PIB"].sum()
```
